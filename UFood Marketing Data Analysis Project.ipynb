Data Cleaning

```python
# Checking for duplicates.

food[food.duplicated()].count()
```

```python
# Dropping the duplicate rows.

food.drop_duplicates(keep= False, inplace= True)
```

```python
# Checking for duplicates to make sure they have been removed.

food[food.duplicated()].count()
```

```python
# Looking at the data's datatypes. Noticed we have no nulls in the data.

food.info()
```

```python
# Creating a new column for total kids in the home. It's much easier to look at than having separate columns for 'Kid' and 'Teen'.

food['total_children'] = food[['Kidhome', 'Teenhome']].sum(axis= 1)
```

```python
# Combining all 5 of the Marital columns into 1. 
# I did this by passing a dictionary that says if the value for the referenced column is 1, then replace with either 5, 4, 3, 2, or 1 for the respective marital status' ({1:5}). 
# If the value is 0 then keep it as 0 ({0:0}).
# Then I summed them all together into a new Martial_Status column where each number represents a particular marital status (ie. Married = 4, Together = 2, etc.)

food['marital_Divorced'] = food['marital_Divorced'].replace({1:5,0:0})
food['marital_Married'] = food['marital_Married'].replace({1:4,0:0})
food['marital_Single'] = food['marital_Single'].replace({1:3,0:0})
food['marital_Together'] = food['marital_Together'].replace({1:2,0:0})
food['marital_Widow'] = food['marital_Widow'].replace({1:1,0:0})

food['Marital_Status'] = food[['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']].sum(axis = 1)
```

```python
# Using the .map() function to assign each number to a string output of 'Divorced', 'Married', 'Single', etc. Much easier to read and use for analysis.

food['Marital_Status_Str'] = food['Marital_Status'].map({5:'Divorced', 4:'Married', 3:'Single', 2:'Together', 1:'Widow'})
```

```python
# Doing the exact same thing as I did with Marital Status with Education Level. Makes it much easier to read and use for analysis.

food['education_2n Cycle'] = food['education_2n Cycle'].replace({1:1,0:0})
food['education_Basic'] = food['education_Basic'].replace({1:2,0:0})
food['education_Graduation'] = food['education_Graduation'].replace({1:3,0:0})
food['education_Master'] = food['education_Master'].replace({1:4,0:0})
food['education_PhD'] = food['education_PhD'].replace({1:5,0:0})

food['Education_Status'] = food[['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']].sum(axis = 1)
```

```python
# Mapping the numeric education values to a string name for ease of reading again.

food['Education_Status_Str'] = food['Education_Status'].map({5:'PhD', 4:'Master', 3:'Graduation', 2:'Basic', 1:'2n Cycle'})
```

```python
# Creating a column to produce either a 1 if the customer accepted any campaign, or a 0 if the customer did not accept any campaigns.

food['Accepted_Campaigns'] = food[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp2', 'AcceptedCmp1']].sum(axis = 1)

food['Accepted_Campaigns'] = (food['Accepted_Campaigns'] != 0).astype(int)
```

Age Segmentation Analysis

```python
# I want to do some analysis on which age groups are accepting the most campaigns. 

# Creating a list of tuples to use as age containers.
age_groups = [(23,30), (31,40), (41,50), (51,60), (61,70), (71,85)]

# A basic function to take the Age of a customer and sort it into their respective age groups.
def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown.")

# Creating a new column and applying the return value of the age group function as the value.
food['Age_Group'] = food['Age'].apply(assign_age_group)
```

```python
#Creating a point plot for the accepted campaign percentage for each age group. Used a list to order from small to large.

age_order = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70', '71 - 85']

sns.pointplot(data= food, x= 'Age_Group', y= 'Accepted_Campaigns', order= age_order)

# This data is probably skewed due to low confidence in the 21-30 and 71-85 age groups.
```

```python
# To check if my hypothesis is true, I'm taking a count of the age groups.

counts = food['Age_Group'].value_counts()
counts

# As expected, the 21-30 and 71-85 age groups have drastically lower numbers than the rest of the groups.
```

```python
# Turning these age numbers indo a percentage of total customers and then creating a dataframe from it.
percentage = counts / food.shape[0]
percent_food = percentage.reset_index()

# Naming the columns of this dataframe.
percent_food.columns = ["age_group", 'percentage']

# Sorting based on age group.
percent_food = percent_food.sort_values('age_group')

percent_food
```

```python
# Creating a bar plot to show the percentage of accepted campaigns per age group.

sns.barplot(x= 'age_group', y= 'percentage', data= percent_food)
plt.title('Percentage of Accepted Campaigns per Age Group')
plt.show()

# Reshowing the pointplot from earlier to compare.
age_order = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70', '71 - 85']

sns.pointplot(data= food, x= 'Age_Group', y= 'Accepted_Campaigns', order= age_order)
plt.show()


# This data tells me that the while the 23-30 and 71-85 age groups make up a much smaller percentage of total sales, they are also the ones accepting the campaigns at higher rates.
# The core audience for accepting campaigns right now is 30 - 70 year olds.
```

```python
# Creating a bar graph to visualize the amount spent per Age Group.

grouped_amounts = food.groupby('Age_Group')['MntTotal'].sum().reset_index()

sns.barplot(x= 'Age_Group', y= 'MntTotal', data= grouped_amounts)
plt.title('Amount Spent per Age Group')
plt.show()

# This follows the same trend we saw previously, ages 30-70 are spending the most money.
```

```python
# Creating a bar graph to visualize the amount spent per Age Group that has accepted a campaign.

# Filtering down to where accepted campaigns is above 0.
accept_campaign = food[food['Accepted_Campaigns'] != 0]

grouped_amounts_accepted = accept_campaign.groupby('Age_Group')['MntTotal'].sum().reset_index()

sns.barplot(x= 'Age_Group', y= 'MntTotal', data= grouped_amounts_accepted)
plt.title('Amount Spent per Age Group')
plt.show()

# When filtering down to show the amount spent per age group that has accepted a campaign, we can see the 23-30 and 71-85 age groups start to look a little closer to the average.
# This could be due to them accepting the campaigns at a higher rate than other age groups.
```
